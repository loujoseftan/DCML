# [Wikipedia](https://blog.lateral.io/2015/06/the-unknown-perils-of-mining-wikipedia/) articles

```python
from scipy.sparse import csr_matrix
```

```python
# load data
articles = pd.read_csv('https://s3.amazonaws.com/assets.datacamp.com/production/course_2141/datasets/wikipedia-vectors.csv', index_col=0)
articles = articles.transpose()

# create csr_matrix
articles_ = csr_matrix(articles)
articles_

# get list of titles
titles = list(articles.index.to_numpy())
```

## Applying NMF to Wikipedia articles

```python
from sklearn.decomposition import NMF
```

```python
# create NMF model
model1 = NMF(n_components=6)
```

```python
# fit to data
model1.fit(articles)
```

```python
# transform articles
nmf_features = model1.transform(articles)
```

```python
nmf_features.shape
```

```python
# create df aligning titles to nmf_features
df = pd.DataFrame(nmf_features, index=titles)
df.head()
```
