```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
import librosa as lr
from glob import glob
```

# Heartbeat Sounds Dataset

```python
# list all wav files in the wav_files folder
audio_files = glob('./wav_files/*.wav')
```

## Inspecting the classification data

```python
# read in the first audio file, create the time array
audio, sfreq = lr.load(audio_files[0]) # sfreq - sampling frequency
time = np.arange(0, len(audio)) / sfreq

# plot audio over time
fig, ax = plt.subplots(figsize=(8, 6))

ax.plot(time, audio)
ax.set(xlabel='Time (s)',
       ylabel='Sound Amplitude')

plt.show()
```

# New York Stock Exchange Dataset

```python
# read in the data
nyse = pd.read_csv('prices_nyse.csv', index_col=0)
print(nyse.shape)
nyse.head()
```

```python
nyse.info()
```

```python
# convert the index of the DataFrame to datetime
nyse.index = pd.to_datetime(nyse.index)

# loop through each column, plot its values over time
fig, ax = plt.subplots(figsize=(8, 6))

for column in nyse.columns:
    nyse[column].plot(ax=ax,
                      label=column)
ax.legend()

plt.show()
```
